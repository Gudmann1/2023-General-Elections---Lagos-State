```python
#Import Relevant Python Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Read Election data breakdown from csv file
Total_vote_cast = pd.read_csv('Elections1.csv')
voting_Brkdown = pd.read_csv('Elections2.csv')
```

```python
#Plot a facetgrid of voting from the Presidential and Governorship across the 20 Local government areas.

g = sns.FacetGrid(data=voting_Brkdown, col = 'LGA', sharex = False, sharey = False, height = 3, 
                  col_wrap = 5)
g.map(sns.barplot, 'Party', 'Count', 'Category', order = ['APC','ADC','LP','PDP'],
     palette = sns.color_palette(['green','red']))
g.add_legend();
```
